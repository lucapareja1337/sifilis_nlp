```python
import seaborn as sns

def pareto(texto, coluna_texto, quantidade):
  todas_palavras = ' '.join([noticias_texto for noticias_texto in texto[coluna_texto]])
  token_espaco = nltk.tokenize.WhitespaceTokenizer()
  token_frase = token_espaco.tokenize(todas_palavras)
  frequencias = nltk.FreqDist(token_frase)
  df_frequencias = pd.DataFrame({'Palavras': list(frequencias.keys()),
                               'Frequencia': list(frequencias.values())})
  df_frequencias = df_frequencias.nlargest(n=quantidade, columns='Frequencia')

  total = df_frequencias['Frequencia'].sum()
  df_frequencias['Porcentagem'] = df_frequencias['Frequencia'].cumsum() / total * 100

  plt.figure(figsize=(12,8))
  ax = sns.barplot(data=df_frequencias, x='Palavras', y='Frequencia', color='gray')
  ax2 = ax.twinx()
  sns.lineplot(data=df_frequencias, x='Palavras', y='Porcentagem', color='red', sort=False, ax=ax2)
  plt.show()



pareto(noticias_texto, "Texto_noticia", 10)
```

```python
#Função para stop words das amostras
def stop_words_amostra(dataset,coluna):
    palavras_irrelevantes = nltk.corpus.stopwords.words("portuguese")
    texto_processado = list()
    for text in dataset[coluna]:
        novo_texto = list()
        text = ''.join([j for j in text if not j.isdigit()])
        palavras_texto = token_espaco.tokenize(str(text))
        for palavra in palavras_texto:
            if palavra not in palavras_irrelevantes:
                novo_texto.append(palavra)
        texto_processado.append(' '.join(novo_texto))
        
    dataset[coluna] = texto_processado
    
```

```python
def grau_similaridade(dataset_1,dataset_2,coluna_1,coluna_2,linha_1,linha_2):
    #Amostra
    texto_origem = dataset_2[coluna_2][linha_2]
    #Noticia
    texto_comparado =dataset_1[coluna_1][linha_1]
    n = 1
    counts = CountVectorizer(analyzer ='word',ngram_range=(n,n))
     #Matriz dos pesos de n_grama
    n_grams = counts.fit_transform([texto_comparado,texto_origem])
    print('\n')
    n_grams.toarray()
    intersec_texto = np.amin(n_grams.toarray(),axis = 0)
    intersec_count = np.sum(intersec_texto)
    index_0 = 0 
    count_index = np.sum(n_grams.toarray()[index_0])
    similaridade_ngram= intersec_count/count_index
    return similaridade_ngram
```

```python
from collections import defaultdict

def rotulos(dataset_1,dataset_2, coluna_1):
    #lines, columns = dataset_1.shape
    linhas,colunas = dataset_2.shape
    rotulo = list()
    print(colunas)
    #print(lines)
    #For para varrer tratamento_4
    for i in  range(len(dataset_1[coluna_1])):
        vetor = []
        
        #For para varrer cada rótulo do dataset amostras
        for j in dataset_2:
            aux = 0 
            #For para varrer cada amostra do dataset amostra
            for x in range(linhas): 
                #print(linhas)      
                #similaridade= grau_similaridade(noticias_texto,amostras,"tratamento_4","Tratamento",0,1)
                texto_origem = dataset_2[j][x]
                #Noticia
                texto_comparado =dataset_1[coluna_1][i]
                n = 1
                counts = CountVectorizer(analyzer ='word',ngram_range=(n,n))
                #Matriz dos pesos de n_grama
                n_grams = counts.fit_transform([texto_comparado,texto_origem])
                #print('\n')
                n_grams.toarray()
                intersec_texto = np.amin(n_grams.toarray(),axis = 0)
                intersec_count = np.sum(intersec_texto)
                index_0 = 0 
                count_index = np.sum(n_grams.toarray()[index_0])
                similaridade = intersec_count/count_index
                #similaridade = intersec_count
                #similaridade = grau_similaridade(dataset_1,dataset_2,coluna_1,j,i,x)
                #print(similaridade)
                if similaridade >= aux:
                    aux = similaridade
                    
            vetor.append(aux)
            #print(vetor)
        rotulo.append(vetor)
    return rotulo 
```
